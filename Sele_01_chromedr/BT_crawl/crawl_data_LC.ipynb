```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python

# 1. Khởi tạo
options = webdriver.ChromeOptions()
# options.add_argument("--start-maximized") # Mở rộng màn hình để dễ cuộn
driver = webdriver.Chrome(options=options)

driver.get("https://nhathuoclongchau.com.vn/thuc-pham-chuc-nang")
print("Đang vào trang web...")
time.sleep(3)

# 2. Vòng lặp: Tìm nút -> Cuộn tới nút -> Bấm
count = 0

while True:
    try:
        # Tìm nút "Xem thêm" (Dùng XPath chứa class và text như cũ)
        button_xpath = "//button[contains(@class, 'mt-3') and contains(., 'Xem thêm')]"
        
        # Đợi tối đa 5s để nút xuất hiện trong DOM
        load_more_btn = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, button_xpath))
        )
        
        # --- ĐÂY LÀ PHẦN CUỘN BẠN YÊU CẦU ---
        # Lệnh JS: "Hãy cuộn sao cho cái nút này nằm ngay GIỮA màn hình"
        # block: 'center' rất quan trọng để nút không bị che bởi header/footer
        driver.execute_script("arguments[0].scrollIntoView({block: 'center', behavior: 'smooth'});", load_more_btn)
        
        # Nghỉ 1 xíu cho cuộn xong hẳn (nhìn cho giống người)
        time.sleep(1)
        
        # Sau khi cuộn đến nơi rồi thì click
        # Dùng click() thường vì giờ nút đã hiện ra rồi
        load_more_btn.click()
        
        count += 1
        print(f"Lần {count}: Đã cuộn và bấm 'Xem thêm'")
        
        # Đợi dữ liệu mới tải về
        time.sleep(10)
        
    except Exception as e:
        # Khi không tìm thấy nút nữa (đã load hết) thì vòng lặp sẽ gãy (break)
        print("\n--- Đã tải hết toàn bộ sản phẩm (Không thấy nút nữa) ---")
        break

# 3. Thu thập dữ liệu sau khi đã mở bung hết trang
print("Đang quét toàn bộ danh sách sản phẩm...")

# Dùng contains để lấy danh sách (bỏ số [9] đi để lấy hết)
# XPath này tìm tất cả thẻ div con nằm trong lưới sản phẩm
product_elements = driver.find_elements(By.XPATH, '//*[@id="category-page__products-section"]/div[2]/div/div')

print(f"Tổng cộng tìm thấy: {len(product_elements)} sản phẩm.")

# 4. Lưu HTML ra file
html_content = "<html><body>"
for item in product_elements:
    html_content += item.get_attribute('outerHTML') + "\n<hr>\n"
html_content += "</body></html>"

file_name = "longchau_scroll_full.html"
with open(file_name, "w", encoding="utf-8") as f:
    f.write(html_content)

print(f"Đã lưu xong file '{file_name}'.")
driver.quit()
```
